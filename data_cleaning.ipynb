**#salary parsing**

```python
df.shape
```

```python
# rule_sal_neg= (df["Salary Estimate"]== '-1')
# df.loc[rule_sal_neg]

rule_sal_not_neg= (df["Salary Estimate"]!= '-1')
df= df.loc[rule_sal_not_neg]
```

```python
df.shape
```

```python
salary= df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
salary_minus_kds= salary.apply(lambda x: x.replace('K', '').replace('$', ''))
salary_minus_kds
```

```python
df['hourly']= df["Salary Estimate"].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided']= df["Salary Estimate"].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df.loc[(df['hourly']==1)]
```

```python
df.loc[(df['employer_provided']==1)].head()
```

```python
minus_texts_in_salary= salary_minus_kds.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:',''))
```

```python
minus_texts_in_salary.head()
```

```python
minus_texts_in_salary.unique()
```

```python
df["min_salary"]= minus_texts_in_salary.apply(lambda x: x.split("-")[0])
df["min_salary"] = df["min_salary"].astype(int)
df["min_salary"].dtype
```

```python
df["max_salary"]= minus_texts_in_salary.apply(lambda x: x.split("-")[1])
df["max_salary"] = df["max_salary"].astype(int)
df["max_salary"].dtype
```

```python
df['avarage_salary']= (df["min_salary"]+df["max_salary"])/2
df['avarage_salary']
```

```python
df["company_text"]= df.apply(lambda x: x['Company Name'] if x['Rating']< 0 else x['Company Name'][:-3], axis=1)
```

```python
df["company_text"]
```

```python
df['job_states']= df["Location"].apply(lambda x: x.split(",")[1])
df.head()
```

```python
df['job_states'].value_counts()
```

```python
df.columns
```

```python
df["same_state"]= df.apply(lambda x: 1 if x["Location"]== x["Headquarters"] else 0, axis=1)
df.head()
```

```python
df["age"]= df["Founded"].apply(lambda x: x if x<1 else 2022-x)
df.head()
```

```python
df["python_req"]= df["Job Description"].apply(lambda x: 1 if "python" in x.lower() else 0)
df["python_req"].value_counts()
```

```python
df["r_req"]= df["Job Description"].apply(lambda x: 1 if "r studio" in x.lower() else 0)
df["r_req"].value_counts()
```

```python
df["spark_req"]= df["Job Description"].apply(lambda x: 1 if "spark" in x.lower() else 0)
df["spark_req"].value_counts()
```

```python
df["excel_req"]= df["Job Description"].apply(lambda x: 1 if "excel" in x.lower() else 0)
df["excel_req"].value_counts()
```

```python
df=df.drop(["Unnamed: 0"], axis=1)
df.head()
```

```python
df.to_csv("salary_data_cleaned.csv", index= False)
```
